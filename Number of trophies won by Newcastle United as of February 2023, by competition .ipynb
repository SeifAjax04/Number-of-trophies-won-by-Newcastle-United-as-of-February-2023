```python
import matplotlib.pyplot as plt

# Data
competitions = ['FA Cup', 'English Football\n Champions', 'English Second-tier\n Champions', 'Inter-Cities\n Fairs Cup', 'FA Charity\n Shield']
trophies = [6, 4, 4, 1, 1]

# Define colors for each trophy
trophy_colors = ['dimgrey', 'black', 'grey', 'silver', 'gainsboro']

# Plotting
plt.figure(figsize=(13, 8))
bars = plt.barh(competitions, trophies, color=trophy_colors)
plt.xlabel('Number of Trophies', fontweight='bold')  # Make x-axis label bold
plt.title('Number of Trophies won by Newcastle United by Competition\n (as of February 2023)', fontweight='bold', color='red', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Make the competitions labels bold
plt.yticks(fontweight='bold')

# Displaying the values on the bars and add labels in boxes
for bar, value in zip(bars, trophies):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, value, ha='left', va='center', bbox=dict(facecolor='white', alpha=0.5, edgecolor='black'))

# Save the plot as a PNG image
plt.savefig('Number of trophies won by Newcastle United as of February 2023.png', dpi=300)

plt.tight_layout()
plt.show()
```
